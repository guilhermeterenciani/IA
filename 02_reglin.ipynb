```python
i = 10
xbatch = x[i:i+batch_size]
ybatch = y[0,i:i+batch_size]
```

```python
for i in range(0,n,batch_size):
    xbatch = x[i:i+batch_size]
    ybatch = y[0,i:i+batch_size]
```

```python
theta_hat = np.matrix([30.0,-2.0])
alpha = 0.01
```

```python
vloss_minilote = []
vminilote = []
for epoch in range(1000):
    for i in range(0,n,batch_size):
        xbatch = x[i:i+batch_size]
        ybatch = y[0,i:i+batch_size]
        loss = np.power(theta_hat@xbatch.T - ybatch,2)
        theta_hat = theta_hat - (alpha*2*(theta_hat@xbatch.T - ybatch)@xbatch)/n
        vloss_minilote.append(np.mean(loss))
        vminilote.append(theta_hat.tolist().copy())
        print(theta_hat)
```

```python
y_hat = theta_hat@x.T

plt.scatter(x[:,1], y.tolist())
plt.scatter(x[:,1], y_hat.tolist())
```

```python
plt.plot(vloss_minilote[:100])
```
